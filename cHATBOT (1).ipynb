```python
import openai
import ipywidgets as widgets
from IPython.display import display, clear_output

# Clave con API_KEY
openai.api_key = 'MYAPI_KEY'

# Prompts
def choose_prompt(user_input):
    base_prompt = """Soy un chatbot de atención al cliente especializado en productos personalizados para un ecommerce. Aquí hay algunas preguntas frecuentes y sus respuestas:

    Cliente: ¿Qué opciones de personalización ofrecen para las remeras?
    Chatbot: Ofrecemos una variedad de opciones como serigrafía, impresión digital directa y bordado. Puede elegir entre diferentes tamaños, colores y diseños.

    Cliente: ¿Puedo solicitar un diseño personalizado para una taza?
    Chatbot: Claro, con gusto aceptamos diseños personalizados. Por favor, envíenos su diseño o descríbanos su idea y nosotros nos encargaremos del resto.

    Cliente: ¿Qué métodos de impresión utilizan para la hidrografía?
    Chatbot: Utilizamos una técnica de inmersión en agua con películas especiales para crear acabados únicos y duraderos en una variedad de superficies.

    Cliente: ¿Tienen descuentos por volumen en pedidos de serigrafía?
    Chatbot: Sí, ofrecemos descuentos por volumen a partir de 20 unidades. Cuanto mayor sea su pedido, mayor será el descuento.

    Cliente: ¿Qué garantía tienen sobre la calidad de impresión?
    Chatbot: Ofrecemos una garantía de satisfacción del 100% en todos nuestros productos. Si no está satisfecho con la calidad de impresión, por favor contáctenos para una solución.

    Cliente: ¿Cómo puedo seguir mi pedido?
    Chatbot: Una vez que su pedido sea enviado, le proporcionaremos un número de seguimiento para que pueda monitorear su entrega en tiempo real.

    Cliente: ¿Qué ocurre si mi pedido llega dañado?
    Chatbot: En caso de que su pedido llegue dañado, por favor contáctenos inmediatamente con fotos del daño para procesar una reposición o un reembolso.

    Cliente: ¿Ofrecen envíos internacionales?
    Chatbot: Sí, realizamos envíos a nivel internacional. Los costos y tiempos de envío pueden variar dependiendo de su ubicación.

    Cliente: ¿Cómo puedo contactarlos para una consulta personalizada?
    Chatbot: Puede contactarnos a través de nuestro formulario en línea, correo electrónico o por teléfono. Nuestro equipo estará encantado de asistirle con su consulta personalizada.

    Cliente: ¿Cuáles son sus horarios de atención?
    Chatbot: Nuestro horario de atención es de lunes a viernes desde las 8:00 a.m hasta las 17:00 horas.

    Cliente: ¿Cómo puedo realizar un pago?
    Chatbot: Puede realizar el pago a través de tarjeta de crédito, PayPal o transferencia bancaria. ¿Necesita ayuda con algún método en particular?

    Cliente: ¿Cuánto tiempo tarda el envío?
    Chatbot: El tiempo de envío promedio es de 3-5 días hábiles dependiendo de su ubicación.


    Cliente: ¿Qué políticas tienen para devoluciones o cambios?
    Chatbot: Tenemos una política de devoluciones de 30 días. Los productos deben estar en su estado original y el cliente es responsable de los costos de envío de la devolución.

    Cliente: {user_input}
    Chatbot: """
    return base_prompt.format(user_input=user_input)



# Función para elegir el prompt adecuado basado en la entrada del usuario
#def choose_prompt(user_input):
#    elif "envío" in user_input.lower() or "entrega" in user_input.lower():
#        return f"{user_input}\n{prompts['envios']}"
#    elif "horario" in user_input.lower() or "atienden" in user_input.lower():
#        return f"{user_input}\n{prompts['horario']}"
#    else:
#        return f"{user_input}\n{prompts['saludo']}"

# Texto de Salida
output_area = widgets.Output()

# Texto de entrada
input_text = widgets.Text(
    value='',
    placeholder='Escribe tu pregunta aquí...',
    description='Pregunta:',
    disabled=False
)

# Función in_out con interfaz de usuario
def get_chatbot_response(change):
    if change['type'] == 'change' and change['name'] == 'value':
        user_input = change['new']
        chosen_prompt = choose_prompt(user_input)
        
        with output_area:
            clear_output()
            try:
                response = openai.Completion.create(
                    engine="text-davinci-003",
                    prompt=chosen_prompt,
                    max_tokens=150
                )
                print(response.choices[0].text.strip())
            except Exception as e:
                print(f"Error al obtener la respuesta de OpenAI: {e}")

# Observa el campo de texto para la entrada (cuando el usuario presiona Enter)
input_text.observe(get_chatbot_response, names='value')

# Mostrar widgets
display(input_text, output_area)
```
